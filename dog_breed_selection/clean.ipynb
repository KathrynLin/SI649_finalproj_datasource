```python
import pandas as pd
```

```python
dog_all=pd.read_csv("best_in_show.csv")
```

```python
dog_all.head()
```

```python
dog_all["Breed"]=dog_all["Dog breed"]
```

```python
dog_height=pd.read_csv("AKC Breed Info.csv")
```

```python
dog_height.head()
```

```python
dog_inte=pd.read_csv("dog_intelligence.csv")
```

```python
dog_inte.head()
```

```python
dog_merge=pd.merge(dog_all,dog_height,on='Breed')
```

```python
dog_merge=pd.merge(dog_merge,dog_inte,on='Breed')
```

```python
dog_merge.info()
```

```python
dog_merge.columns
```

```python
dog_merge["height_high_inches"] = pd.to_numeric(dog_merge["height_high_inches"], errors='coerce')
dog_merge["height_low_inches"] = pd.to_numeric(dog_merge["height_low_inches"], errors='coerce')

# Calculate average height
dog_merge['Average_Height'] = (dog_merge["height_high_inches"] + dog_merge["height_low_inches"]) / 2
```

```python
# Convert non-numeric values to NaN
dog_merge["weight_high_lbs"] = pd.to_numeric(dog_merge["weight_high_lbs"], errors='coerce')
dog_merge["weight_low_lbs"] = pd.to_numeric(dog_merge["weight_low_lbs"], errors='coerce')

# Calculate average weight
dog_merge['Average_Weight'] = (dog_merge["weight_high_lbs"] + dog_merge["weight_low_lbs"]) / 2
```

```python
dog_merge.head()
```

```python
dog_filtered=dog_merge[['Breed','datadog score','POPULARITY IN US.1','LONGEVITY','LIFETIME COST, $','Average_Height','Average_Weight','obey','intelligence category']]
```

```python
dog_filtered
```

```python
dog_filtered_cleaned=dog_filtered.dropna()
```

```python
dog_filtered_cleaned
```

```python
intell=dog_filtered_cleaned['intelligence category']
```

```python
intell.unique()
```

```python
mapping = {'Brightest': 5, 'Excellent': 4, 'Above average': 3, 'Average': 2, 'Fair': 1}

# Replace the categories with numbers
dog_filtered_cleaned['intelligence category'] = dog_filtered_cleaned['intelligence category'].replace(mapping)

# Display the DataFrame with replaced values
```

```python
dog_filtered_cleaned = dog_filtered_cleaned.rename(columns={'intelligence category': 'Intelligence'})
dog_filtered_cleaned = dog_filtered_cleaned.rename(columns={'LIFETIME COST, $': 'Lifetime_Cost'})
dog_filtered_cleaned = dog_filtered_cleaned.rename(columns={'POPULARITY IN US.1': 'Popularity'})
dog_filtered_cleaned['Lifetime_Cost'] = dog_filtered_cleaned['Lifetime_Cost'].str.replace('$', '')
```

```python
dog_filtered_cleaned
```

```python
dog_filtered_cleaned.to_csv('dog_filtered_cleaned.csv', index=False)
```

```python
# Convert 'LONGEVITY' column to numeric values
dog_filtered_cleaned['LONGEVITY'] = dog_filtered_cleaned['LONGEVITY'].str.replace('%', '')
dog_filtered_cleaned['LONGEVITY'] = dog_filtered_cleaned['LONGEVITY'].astype(float) / 100



# Min-Max scaling
df_normalized = dog_filtered_cleaned.copy()
for column in df_normalized.columns:
    if column != 'Breed':
        if df_normalized[column].dtype != 'object':
            df_normalized[column] = (df_normalized[column] - df_normalized[column].min()) / (df_normalized[column].max() - df_normalized[column].min())

# Display the normalized DataFrame
df_normalized.head()
```
