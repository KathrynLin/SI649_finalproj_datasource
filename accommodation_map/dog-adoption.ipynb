### Seeing which states have the most imports of dogs and what breeds/types those are ? 

```python
count_of_dogs = 0
contact_country_dict = {}
for i in dog_desc_df['contact_country']:
    if i =='US' : 
        contact_country_dict['breed'] = dog_desc_df['breed_primary']
        contact_country_dict['type.x'] = dog_desc_df['type.x']
        contact_country_dict['contact_state'] = dog_desc_df['contact_state']
        count_of_dogs += 1 
```

```python
contact_country_df = pd.DataFrame(contact_country_dict)
contact_country_df.head()
```

```python
contact_country_df.to_csv('contact_country.csv')
```

```python
df1_dict = {}
count = 0
for i in dog_desc_df['status']:
    if i =='adoptable' : 
        df1_dict['breed'] = dog_desc_df['breed_primary']
        df1_dict['type.x'] = dog_desc_df['type.x']
        count += 1 


adopted_dogs_in_us = pd.DataFrame(df1_dict)

adopted_dogs_in_us.head()
```
