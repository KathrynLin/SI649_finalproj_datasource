```python
import pandas as pd
import re

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/linfangqing/Desktop/dog_economic/data-Y7Cd3.csv')

# Convert currency and percentage fields to numeric types
# Remove '$', ',', and '%' and convert the strings to floats
currency_fields = ['Median Household Income', 'Vet Care, Pet Insurance & Boarding_1', 'Day Care_1']
percentage_fields = ['Vet Care, Pet Insurance & Boarding', 'Day Care']

# Remove currency formatting and convert to float
for field in currency_fields:
    df[field] = df[field].replace('\$', '', regex=True)
    df[field] = df[field].replace(',', '', regex=True)
    df[field] = df[field].replace('%', '', regex=True).astype(float)

# Remove percentage formatting and convert to float
for field in percentage_fields:
    df[field] = df[field].replace('%', '', regex=True).astype(float) / 100  # divide by 100 to convert to a proportion

df['City '] = df['City '].str.extract(r'^(.*?)\s*\^')

# Save the cleaned DataFrame to a new CSV file
clean_csv_file_path = 'cleaned_pet_data.csv'
df.to_csv(clean_csv_file_path, index=False)
df.head()
```
